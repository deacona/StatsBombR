```r
library(StatsBombR)
```

```r
FreeCompetitions()
```

```r
comp <- FreeCompetitions()
# head(comp, n=5)
comp[comp$competition_id == 11, ]
```

```r
comp$season_name[comp$competition_id == 11]
```

```r
matches <- FreeMatches(comp[comp$competition_id == 11, ])
options(repr.matrix.max.rows=60, repr.matrix.max.cols=40)
head(matches, n=5)
```

```r
hist(matches$home_score, n=10)
```

```r
matches["total_goals"] <- matches$home_score + matches$away_score
hist(matches$total_goals, n=10)
```

```r
library(stringr)
# matches["kick_off_hour"] <- stringr::str_sub(matches[,3], start = 1, end = 2)
matches["kick_off_hour"] <- as.numeric(stringr::str_sub(matches$kick_off, start = 1, end = 2))
head(matches$kick_off_hour, n=5)
```

```r
hist(matches$kick_off_hour, breaks=10)
```

```r
lineups <- StatsBombFreeLineups(matches)
head(lineups, n=5)
```

```r
singlematch <- get.matchFree(matches[1,])
nrow(singlematch)
ncol(singlematch)
```

```r
singlematchclean <- allclean(singlematch)
nrow(singlematchclean)
ncol(singlematchclean)
```

```r
options(repr.matrix.max.rows=100, repr.matrix.max.cols=160)
tail(head(singlematchclean, n=100), n=5)
```

```r
shots <- singlematchclean[singlematchclean$type.name == "Shot", ]
shotsinfo <- shots[,c("period", "timestamp", "minute", "play_pattern.name",
                      "position.name", "under_pressure",
                    "team.name", "player.name", "location", "shot.type.name", 
                      "shot.outcome.name", "shot.statsbomb_xg")]
head(shotsinfo, n=10)
```

```r
plot(shotsinfo$minute, shotsinfo$shot.statsbomb_xg)
```

```r
cor(shotsinfo$minute, shotsinfo$shot.statsbomb_xg)
```
